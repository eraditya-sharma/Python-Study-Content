# Logical Operators:
Logical Operators are used to combine two or more conditions and perform the logical operations using Logical AND, Logical OR and Logical NOT

# AND

```python
age=29
if age < 33 and age > 20:  #It will return true when both conditions are true
	print ("Young Man")
else:
	print(" Not Eligible ")
```

# OR

```python
age=29
if age < 18 or age > 60:  #It will returns true when at-least one of the condition is true
	print(" Not Eligible to Work ")
else:
	print(" Please forward Your Resume ")
```

```python
inp1 = int(input('Enter the lower range: '))
inp2 = int(input('Enter the upper range: '))

print("Prime numbers between ",inp1," and ",inp2," are: ")

for num in range(inp1,inp2 + 1):
   # prime numbers are greater than 1
   if num > 1:
       for i in range(2,num):
           if (num % i) == 0:
               break
       else:
           print(num)
```
